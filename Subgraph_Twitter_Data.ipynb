```
import pandas as pd
import matplotlib.pyplot as plt
```

```
data=pd.read_csv('https://raw.githubusercontent.com/shahriariit/opendataset/master/subgraph.csv')
```

```
data[0:8]
```

```
def xValueLoad(x,y):
   vals = [] 
   for i in range(len(x.columns)):
      val = x.iloc[0:y,i] 
      vals.append(val)
    
   vals.remove(vals[0])
   return vals;
```

```
x_list=[]

x_list.append(xValueLoad(data[0:8],8))
x_list.append(xValueLoad(data[10:18],8))
x_list.append(xValueLoad(data[20:28],8))
x_list.append(xValueLoad(data[30:38],8))
x_list.append(xValueLoad(data[40:48],8))
x_list.append(xValueLoad(data[50:58],8))
x_list.append(xValueLoad(data[60:68],8))
x_list.append(xValueLoad(data[70:78],8))
x_list.append(xValueLoad(data[80:88],8))
```

```
y_list=['DT','GB','KNN','LR','MLP','RF','SVM','XGB']
```

```
SMALL_SIZE = 15

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SMALL_SIZE)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=15)    # legend fontsize
plt.rc('figure', titlesize=SMALL_SIZE)   # fontsize of the figure title
```

```
line_color=['green','red','blue']
```

```
def graphFunc(l1,y,l2,lt,col,bboxanch):
  plt.figure(figsize=(13,8))
    
  for i,j,k in zip(l1,l2,l2):
     plt.plot(y,i, color=j, linestyle='solid', linewidth = 3, marker='o', markerfacecolor=k, markersize=12)
  
  plt.xlabel('Classifiers') 
  plt.ylabel('Results') 
  plt.legend(('Average', 'Average of Existing Models', 'Average of Proposed Models'),loc=lt,ncol=col, bbox_to_anchor=bboxanch)   
  plt.show()
```

```
graphFunc(x_list[0],y_list,line_color,'center left',1, (0,0.2,0,0))
```
